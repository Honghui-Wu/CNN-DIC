```python
import pandas as pd

# Define the file path
file_path = "ABAQUS_Formatted_Files_16dpts\Displacement3.2_all_formatted.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head(10)
```

```python
df["X_difference"] = df["X (deformed)"] - df["X (original)"]
df["Y_difference"] = df["Y (deformed)"] - df["Y (original)"]


# Display the first few rows
print(df[["X_difference", "Y_difference"]].head(10))
```

```python
num_data_points = df.shape[0]

num_data_points
```

```python
# Group by 'X (original)'
grouped_df_X = df.groupby("X (original)").agg(["count"])

# Display the grouped DataFrame
print(grouped_df_X)
```

```python
# Group by 'Y (original)' and calculate statistics
grouped_df_Y = df.groupby("Y (original)").agg(["count"])

# Display the grouped DataFrame
print(grouped_df_Y)
```

```python
'''Plot the nodes of the data points, where X(original) and Y(original) are the x and y coordinates, respectively.'''
import matplotlib.pyplot as plt

# Scatter plot of original coordinates
plt.figure(figsize=(8, 6))
plt.scatter(df["X (original)"], df["Y (original)"], marker='o', alpha=0.7)

# Labels and title
plt.xlabel("X (original)")
plt.ylabel("Y (original)")
plt.title("Scatter Plot of Original Nodes")
plt.grid(True)

# Show the plot
plt.show()

```

```python
'''Plot the nodes of the data points, where X(deformed) and Y(deformed) are the x and y coordinates, respectively.'''
import matplotlib.pyplot as plt

# Scatter plot of original coordinates
plt.figure(figsize=(8, 6))
plt.scatter(df["X (deformed)"], df["Y (deformed)"], marker='o', alpha=0.7, color='r')

# Labels and title
plt.xlabel("X (deformed)")
plt.ylabel("Y (deformed)")
plt.title("Scatter Plot of Deformed Nodes")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Scatter plot of original and deformed nodes on the same plot
plt.figure(figsize=(8, 6))

# Original nodes
plt.scatter(df["X (original)"], df["Y (original)"], marker='o', alpha=0.7, label="Original Nodes")

# Deformed nodes
plt.scatter(df["X (deformed)"], df["Y (deformed)"], marker='o', alpha=0.5, label="Deformed Nodes", color='r')

# Labels and title
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Comparison of Original and Deformed Nodes")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
