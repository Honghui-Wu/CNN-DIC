```python
'''Plot the nodes of the data points, where X(original) and Y(original) are the x and y coordinates, respectively.'''
import numpy as np
import matplotlib.pyplot as plt

# Get unique sorted values for X and Y
x_unique = np.sort(df["X (original)"].unique())
y_unique = np.sort(df["Y (original)"].unique())


# Downsample every 10th value
x_sampled = x_unique[::16]
y_sampled = y_unique[::16]

# Filter the dataframe based on the downsampled grid
df_downsampled = df[df["X (original)"].isin(x_sampled) & df["Y (original)"].isin(y_sampled)]

# Plot the downsampled scatter
plt.figure(figsize=(12, 10))
plt.scatter(df_downsampled["X (original)"], df_downsampled["Y (original)"], 
            s=5, marker='o', alpha=0.5)

plt.xlabel("X (original)")
plt.ylabel("Y (original)")
plt.title("Downsampled Scatter Plot of Original Nodes")
plt.grid(True)
plt.axis("equal") 
plt.show()
```

```python
'''Plot deformed nodes'''

# Get unique sorted values for X and Y
x_unique = np.sort(df["X (deformed)"].unique())
y_unique = np.sort(df["Y (deformed)"].unique())

# Downsample every 10th value
x_sampled = x_unique[::16]
y_sampled = y_unique[::16]

# Filter the dataframe based on the downsampled grid
df_downsampled = df[df["X (deformed)"].isin(x_sampled) & df["Y (deformed)"].isin(y_sampled)]

# Plot the downsampled scatter
plt.figure(figsize=(12, 10))
plt.scatter(df_downsampled["X (deformed)"], df_downsampled["Y (deformed)"], 
            s=5, marker='o', alpha=0.5, color='r')

plt.xlabel("X (deformed)")
plt.ylabel("Y (deformed)")
plt.title("Downsampled Scatter Plot of Deformed Nodes")
plt.grid(True)
plt.axis("equal") 
plt.show()
```

EDA for Large Dataset of Displacement/Strain (160 files, 4.4GB)

```python
'''Load data'''

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
from tqdm import tqdm

# === Configuration ===
data_folder = "ABAQUS_Formatted_Files_512x512_160"  # folder path
max_files = 160  # Adjust if testing on fewer files

# === Helper Function ===
def load_data(files):
    dfs = []
    for file in tqdm(files, desc="Loading files"):
        df = pd.read_csv(file)
        df["filename"] = os.path.basename(file)
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# === Load and Sample Files ===
all_files = sorted(glob(os.path.join(data_folder, "Disp*_extracted_formatted.csv")))[:max_files] # all_files is a list
data = load_data(all_files) # data is a Pandas DataFrame

print("Loaded files:", len(all_files))

data.head()
```

```python
data.describe()
```

```python
# === Plot Displacement Histograms ===
plt.figure(figsize=(12, 6))
sns.histplot(data["U1"], bins=100, kde=True, label="U1", color="blue")
sns.histplot(data["U2"], bins=100, kde=True, label="U2", color="green")
plt.title("Histogram of Displacement Components (U1, U2)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ===Plot Strain Magnitude, E11, E22, E12 ===
def compute_strain_magnitude(df):
    return np.sqrt(df["E11"]**2 + df["E22"]**2 + 2 * df["E12"]**2)

data["strain_mag"] = compute_strain_magnitude(data)

plt.figure(figsize=(10, 6))
sns.histplot(data["strain_mag"], bins=100, kde=True, label="strain_mag", color="red")
plt.title("Histogram of Strain Magnitude")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["E11"], bins=100, kde=True, label="E11", color="green")
sns.histplot(data["E22"], bins=100, kde=True, label="E22", color="blue")
sns.histplot(data["E12"], bins=100, kde=True, label="E12", color="purple")
plt.title("Histogram of Strain E11, E22, E12")
plt.grid(True)
plt.legend()
plt.show()
```
